```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn
```

```python
def set_plt_params():
    title_font = {'size':'30', 'color':'black', 'weight':'bold',
                  'verticalalignment':'bottom'} 
    axis_font = {'size':'30'}
    plt.figure(figsize=(10, 10))
    plt.grid(False)
    plt.axes(frameon = 0)
    #plt.tick_params(labelsize=60)
    #plt.ylim([0.76, 0.86])
    #plt.tick_params(labelsize=30)
    seaborn.set_style("whitegrid", {'grid' : False})
```
