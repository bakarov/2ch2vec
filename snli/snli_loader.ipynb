```python
from pandas import DataFrame
import numpy as np
```

```python
def read_data(filename):
    data = [l.strip().split('\t') for l in open(filename)]

    sentence1 = []
    sentence2 = []
    label = []

    for i in data[1:]:
        sentence1.append(i[5])
        sentence2.append(i[6])
        label.append(i[0])

    df = DataFrame()
    df['sentence1'] = sentence1
    df['sentence2'] = sentence2
    df['label'] = label
    return df
```

```python
df_test = read_data('snli_1.0_test.txt')
```
